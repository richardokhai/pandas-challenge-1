## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the subtotal 
# for each line using the unit_price and the qty
df['line_subtotal'] = df['unit_price'] * df['qty']
#Show first two rows with three columns of client 
# with the most highest purchase based on order quantity
# Note: Include - unit_price, qty, and line_subtotal
df[['unit_price', 'qty', 'line_subtotal']].head(2)
```

```python
# Create a column for shipping price.
df['total_weight'] = df['unit_weight'] * df['qty']
# Assume a shipping price of $7 per pound for orders over 50 pounds 
# and $10 per pound for items 50 pounds or under.
#think of lambda + expression (limited expression)
df['shipping_price'] = df['total_weight'].apply(lambda a: 7*a 
                                                if a > 50 
                                                else 10*a)
#show first 3 customers with the highest purchase 
#for unit_price, unit_weight, qty, total_weigh, shipping_price 
df[['unit_price','unit_weight','qty','total_weight','shipping_price']].head(3)

```

```python
# Create a column for the total price using the subtotal 
# and the shipping price along with a sales tax of 9.25%
df['line_subtotal']= df['unit_price'] *df['qty']
df['total_weight'] = df['unit_weight'] * df['qty']
df['shipping_price'] = df['total_weight'].apply(lambda a: 7 * a 
                                                if a >50
                                                else 10 * a)
df['line_price'] = df['line_subtotal'] + df['shipping_price']
df['line_price'] *= 1.0925
#show line subtotal: Shipping_price, line_price for first 3 rows
df[['line_subtotal','shipping_price','line_price']].head(3).round(2)
```

```python
# Create a column for the cost of each line using unit cost, qty, and
# shipping price (assume the shipping cost is exactly 
# what is charged to the client).
df['line_subtotal'] = df['unit_price']*df['qty']
df['total_weight'] = df['unit_weight']*df['qty']
df['shipping_price'] = df['total_weight'].apply(lambda a: 7*a
                                                if a > 50
                                                else 10 *a)
df['line_cost'] = (df['unit_cost']*df['qty'])+df['shipping_price']
df.head(3)
```

```python
# Create a column for profit of each line using line cost and line price
df['line_subtotal'] = df['unit_price'] * df['qty']
df['total_weight'] = df['unit_weight'] * df['qty']
df['shipping_price'] = df['total_weight'].apply(lambda z:7*z
                                                if z > 50 
                                                else 10*z)
df['line_cost'] = (df['unit_cost'])*df['qty'] + df['shipping_price']
df['line_price'] = (df['line_subtotal'] +df['shipping_price'])
df['line_price'] *= 1.0925
df['line_profit'] = df['line_price'] - df['line_cost']
df.head(3)


```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity spend? Check your work from Part 1 for client ids.
t5_clients = df.groupby('client_id')['line_price'].sum().nlargest(5)
client_spending = {}
for client_id in t5_clients.index:
    client_data = df[df['client_id'] == client_id]
    total_spent = client_data['line_price'].sum()
    client_spending[client_id] = total_spent
for client_id, total_spent in client_spending.items():
    print(f"{client_id} : ${total_spent:.2f}")
```

```python
# Create a summary DataFrame showing the totals for the for the top 5 clients with the following information:
# total units purchased, total shipping price, total revenue, and total profit. 
sum_df = df.groupby('client_id').agg(
    total_units_purchased=('qty', 'sum'),
    total_shipping_price=('shipping_price', 'sum'),
    total_revenue=('line_price', 'sum'),
    total_cost=('line_cost', 'sum'),
    total_profit=('line_profit', 'sum')
)

t5_clients = sum_df.nlargest(5, 'total_units_purchased')

t5_clients.head(5)
```

```python
# Format the data and rename the columns to names suitable for presentation.

# Define the money columns. 

# Define a function that converts a dollar amount to millions.


# Apply the currency_format_millions function to only the money columns. 


# Rename the columns to reflect the change in the money format. 
t_5_clients = sum_df.nlargest(5, 'total_units_purchased')


money_columns = ['total_shipping_price', 'total_revenue', 'total_profit', 'total_cost']  # Add 'total_cost' if it exists


def currency_format_millions(val):
    return val / 1_000_000


for mon_col in money_columns:
    t_5_clients[mon_col] = t_5_clients[mon_col].map(currency_format_millions)


t_5_clients.rename(columns={
    'total_units_purchased': 'Units',
    'total_shipping_price': 'Shipping (millions)',
    'total_revenue': 'Total Revenue (millions)',
    'total_profit': 'Total Profit (millions)',
    'total_cost': 'Total Cost (millions)'  
}, inplace=True)


t_5_clients.head(5)
```

```python
# Sort the updated data by "Total Profit (millions)" form highest to lowest and assign the sort to a new DatFrame.
sort_t_5_clients = t_5_clients.sort_values(by='Total Profit (millions)', ascending=False)
sort_t_5_clients.head(5)
```
